Create ECDF graph and Quantile-Quantile graph

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create ECDF plot
plt.figure(figsize=(8, 5))
sns.ecdfplot(data=tips, x="total_bill")

# Add labels and title
plt.title("ECDF of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("ECDF")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate synthetic normally distributed data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

# Create Q-Q plot
plt.figure(figsize=(8, 6))
stats.probplot(data, dist="norm", plot=plt)

# Add title
plt.title("Q-Q Plot: Normal Distribution Check")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Load real dataset
tips = sns.load_dataset("tips")
data = tips["total_bill"]

# Q-Q plot for 'total_bill'
plt.figure(figsize=(8, 6))
stats.probplot(data, dist="norm", plot=plt)
plt.title("Q-Q Plot of Total Bill (Seaborn 'tips')")
plt.grid(True)
plt.show()
```
