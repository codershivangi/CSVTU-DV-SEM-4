```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 2 * x + 1 + np.random.normal(scale=3, size=100)  # Linear with noise

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# Create a scatter plot with regression line and confidence band
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.regplot(x='x', y='y', data=data, ci=95, line_kws={"color": "red"}, scatter_kws={"s": 30})
plt.title("Trend Line with 95% Confidence Band")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
