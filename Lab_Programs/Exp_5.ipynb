To show an example of skewed data and removal of skewedness

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, boxcox

# 1. Create right-skewed data (exponential distribution)
np.random.seed(42)
skewed_data = np.random.exponential(scale=2, size=1000)

# Create DataFrame
df = pd.DataFrame({'Original': skewed_data})

# 2. Plot original data distribution
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Original'], kde=True, color='orange')
plt.title(f"Original Skewed Data (Skew = {skew(df['Original']):.2f})")

# 3. Apply log transformation (add small constant to avoid log(0))
df['Log Transformed'] = np.log1p(df['Original'])  # log(1 + x)

plt.subplot(1, 2, 2)
sns.histplot(df['Log Transformed'], kde=True, color='green')
plt.title(f"Log-Transformed Data (Skew = {skew(df['Log Transformed']):.2f})")

plt.tight_layout()
plt.show()
```
