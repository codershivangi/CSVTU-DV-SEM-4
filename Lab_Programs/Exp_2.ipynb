```python
# Experiment - 2
# Use different color scales on the Rainfall Prediction dataset.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example of a rainfall prediction dataset (10X10 matrix for simplicity)
# You can replace this with your actual dataset
data = {
    'Location_1': [1.2, 2.3, 3.1, 0.5, 1.2, 0.7, 3.0, 2.4, 1.8, 2.1],
    'Location_2': [2.3, 1.9, 2.2, 3.8, 3.2, 1.5, 1.0, 3.5, 1.4, 1.1],
    'Location_3': [3.0, 2.4, 1.5, 2.9, 2.7, 3.1, 0.1, 2.6, 2.0, 2.5],
    'Location_4': [1.1, 3.3, 3.4, 1.9, 1.4, 1.9, 2.5, 3.3, 1.6, 1.7],
    'Location_5': [2.2, 1.7, 1.6, 3.5, 3.2, 2.8, 3.1, 1.9, 3.2, 2.3],
    'Location_6': [2.1, 1.4, 2.3, 1.3, 1.1, 2.4, 2.5, 1.7, 3.0, 2.1],
    'Location_7': [1.5, 2.8, 2.3, 1.6, 3.0, 1.7, 2.1, 2.2, 1.9, 3.1],
    'Location_8': [1.2, 3.0, 1.6, 2.1, 2.4, 1.9, 2.8, 3.2, 1.8, 2.1],
    'Location_9': [2.3, 2.0, 2.8, 2.5, 3.2, 1.6, 2.9, 1.4, 2.1, 2.7],
    'Location_10': [1.8, 2.7, 3.2, 1.2, 2.1, 2.9, 1.8, 2.3, 3.0, 1.5]
}

# Convert this into a pandas DataFrame
df = pd.DataFrame(data)

# Plotting the heatmap
plt.figure(figsize=(10, 8))

# Using Seaborn heatmap to display the rainfall data with a custom color scale
sns.heatmap(df, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)

# Adding title and labels
plt.title('Rainfall Prediction Heatmap')
plt.xlabel('Location')
plt.ylabel('Time')

# Show the plot
plt.show()
```
